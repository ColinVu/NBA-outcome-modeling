```python
from basketball_reference_web_scraper import client
from basketball_reference_web_scraper.data import Team
import pandas as pd
import time


schedule = client.season_schedule(season_end_year=2021)


processed_dates = set()
all_games = []

for game in schedule:
    game_date = game['start_time'].date()
    if game_date in processed_dates:
        continue

    try:

        daily_games = client.team_box_scores(
            day=game_date.day,
            month=game_date.month,
            year=game_date.year
        )


        for game_data in daily_games:
            game_data['season'] = 2021
            game_data['game_date'] = game_date

        all_games.extend(daily_games)
        processed_dates.add(game_date)


        time.sleep(3)

    except Exception as e:
        print(f"Error fetching {game_date}: {e}")


df = pd.DataFrame(all_games).drop_duplicates()
df.to_csv('2021_nba_game_stats.csv', index=False)
```
